```
import cv2 # Membuka alat membaca gambar dan memberi nama alatnya "cv2"
import numpy as np # Membuka alat matematika dan memberi nama alatnya "numpy" atau "np"
import matplotlib.pyplot as plt # Membuka alat menunjukkan gambar dan memberi nama alatnya "matplotlib.pyplot" atau "plt"
%matplotlib inline # Mengaktifkan mode inline plotting untuk Matplotlib di dalam notebook.

img_path = 'Dila.jpg' # Menyimpan path atau lokasi file gambar dengan nama 'Dila.jpg'.
img = cv2.imread(img_path) # Memberi perintah pada komputer untuk membaca gambar dengan alat "cv2" dan memberi nama gambar tersebut "img"
print(img.shape) # Menampilkan bentuk atau ukuran gambar

plt.imshow(img)  # Menampilkan gambar

fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Mengubah mode warna gambar dari BGR ke RGB
plt.imshow(fix_img)  # Menampilkan gambar yang telah diubah mode warnanya

R, G, B = fix_img[:,:,0], fix_img [:,:,1], fix_img [:,:,2] # Membagi gambar menjadi tiga saluran warna: Merah (R), Hijau (G), dan Biru (B).
print(np.array(fix_img))  # Menampilkan matriks (array dua dimensi) yang merepresentasikan gambar.

```

```
fix_img[:] = np.max(fix_img, axis = -1, keepdims = 1)/2 + np.min(fix_img, axis = -1, keepdims = 1)/2
# Memberi perintah kepada komputer untuk melakukan sesuatu pada gambar yang sudah diubah warnanya
# Kode ini menghitung suatu nilai untuk setiap piksel di gambar

print(np.array(fix_img[:])) # Menunjukkan  nilai-nilai yang sudah dihitung
plt.axis('off') # Menghilangkan garis tepi pada gambar
plt.imshow(fix_img[:], cmap ='gray') # Menunjukkan gambar hitam putih
plt.savefig('metode lightness.jpg') # Menyimpan gambar yang sudah diubah
```

```
gray_img = np.mean(fix_img, axis = 2)
# Memberi perintah kepada komputer untuk menghitung rata-rata warna di setiap piksel di gambar yang sudah diubah warnanya

print(np.array(gray_img)) # Meminta komputer menunjukkan hasil rata-rata yang sudah dihitung

plt.axis('off') # Menghilangkan garis tepi pada gambar
plt.imshow(gray_img, cmap = 'gray') # Memberi perintah kepada komputer untuk menunjukkan gambar yang sudah dihitung rata-ratanya dengan warna abu-abu.
plt.savefig('Metode Average.jpg', bbox_inches = 'tight')
# Memerintahkan komputer untuk menyimpan hasilnya menjadi gambar baru dengan nama 'Metode Average.jpg'
# Menyuruh agar garis putih di sekitar gambar dihilangkan
```

```
lumi_img = (0.2126*R) + (0.7152*G) + (0.0722*B)
# Memberi perintah kepada komputer untuk menghitung suatu nilai untuk setiap piksel di gambar yang sudah diubah warnanya
# Nilai ini dihitung dengan cara tertentu menggunakan rumus matematika

print(lumi_img) # Meminta komputer menunjukkan hasil nilai yang sudah dihitung

plt.axis('off') # Menghilangkan garis tepi pada gambar
plt.imshow(lumi_img,cmap ='gray') # Memberi perintah kepada komputer untuk menunjukkan gambar yang sudah dihitung nilai tersebut dengan warna abu-abu
plt.savefig('Metode Average.jpg', bbox_inches = 'tight')
# Memerintahkan komputer untuk menyimpan hasilnya menjadi gambar baru dengan nama 'Metode Average.jpg'
# Menyuruh agar garis putih di sekitar gambar dihilangkan
```

```
wavr_img = (0.299*R) + (0.587*G) + (0.114*B)
# Memberi perintah kepada komputer untuk menghitung suatu nilai untuk setiap piksel di gambar yang sudah diubah warnanya.
# Nilai ini dihitung dengan cara tertentu menggunakan rumus matematika.

print(wavr_img) # Meminta komputer menunjukkan hasil nilai yang sudah dihitung

plt.axis('off') # Menghilangkan garis tepi pada gambar
plt.imshow(wavr_img, cmap = 'gray') # Memberi perintah kepada komputer untuk menunjukkan gambar yang sudah dihitung nilai tersebut dengan warna abu-abu
plt.savefig('Metode Weighted Average.png', bbox_inches = 'tight')
# Menyuruh komputer untuk menyimpan hasilnya sebagai gambar baru dengan nama 'Metode Weighted Average.png'
```
